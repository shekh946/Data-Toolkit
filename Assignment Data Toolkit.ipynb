**Q6.  What are the differences between NumPy arrays and Python lists?**

 -  Differences Between NumPy Arrays and Python Lists   
 1.  Data Type Homogeneity:
 - NumPy Arrays: Homogeneous, meaning all elements must be of the same data type. This ensures efficient memory usage and faster computations.
 -  Python Lists: Heterogeneous, allowing elements of different data types (e.g., integers, strings, floats) in the same list.
 2. Performance:

 -  NumPy Arrays: Optimized for numerical computations, making them significantly faster for large datasets due to their compact memory layout and vectorized operations.
 -  Python Lists: Slower for numerical operations as they store references to objects, requiring additional overhead.
 3. Memory Efficiency:

 -  NumPy Arrays: More memory-efficient because they store data in contiguous blocks of memory.
 -  Python Lists: Less memory-efficient as they store references to objects, which can lead to fragmentation.
 4. Functionality:

 -  NumPy Arrays: Provide a wide range of mathematical, statistical, and linear algebra operations directly, making them ideal for scientific computing.
 -  Python Lists: Lack built-in numerical operations, requiring manual implementation or external libraries for such tasks.
 5. Flexibility:

 -  NumPy Arrays: Less flexible for general-purpose programming due to their homogeneity constraint.
 -  Python Lists: More flexible for storing and manipulating mixed data types.

**Q15.  How does NumPy handle multidimensional arrays?**

 -  NumPy is a powerful library in Python that efficiently handles multidimensional arrays, which are central to numerical computations. Here's a concise explanation of how it works:

 - Efficient Storage and Operations:  
  NumPy arrays (ndarray) store data in a contiguous block of memory, ensuring efficient storage and faster access compared to Python lists.
Multidimensional arrays are essentially nested arrays, where each dimension is represented as a level of nesting.  
 -  Indexing and Slicing:  
 NumPy allows intuitive indexing and slicing to access or modify specific elements, rows, columns, or subarrays.  
 For example:  
 import numpy as np  
arr = np.array([[1, 2, 3], [4, 5, 6]])  
print(arr[1, 2])  # Access element at row 1, column 2 (output: 6)   
print(arr[:, 1])  # Access all rows, column 1 (output: [2, 5])  
 -  Broadcasting:  
 NumPy supports broadcasting, enabling operations between arrays of different shapes by automatically expanding dimensions when possible.  For example:  
 arr = np.array([[1, 2], [3, 4]])  
print(arr + 10)  # Adds 10 to every element  
 - Reshaping and Transposing:  
 Arrays can be reshaped or transposed to change their dimensions without altering the data:  
 arr = np.array([1, 2, 3, 4])  
reshaped = arr.reshape(2, 2)  # Converts to 2x2 array
transposed = reshaped.T       # Transposes the array
 -  Vectorized Operations:  
 NumPy performs element-wise operations on arrays without explicit loops, making computations faster and more concise:  
arr1 = np.array([[1, 2], [3, 4]])  
arr2 = np.array([[5, 6], [7, 8]])  
print(arr1 + arr2)  # Element-wise addition  
 -  Support for Mathematical Functions:  
 NumPy provides a wide range of mathematical functions (e.g., sin, cos, dot, etc.) that work seamlessly with multidimensional arrays.
By combining these features, NumPy makes handling multidimensional arrays intuitive, efficient, and highly versatile for tasks like data analysis, machine learning, and scientific computing.

**Q17. Explain the difference between apply() and map() in Pandas?**

 -  In Pandas, both apply and applymap are used to apply functions to data, but they differ in their scope and usage. Understanding their distinctions is crucial for efficient data manipulation.

 - apply Method  
 The apply method can be used on both Series and DataFrames. It allows you to apply a function along a specific axis (rows or columns) of a DataFrame or to each element of a Series.  
 Example with DataFrame:  
import pandas as pd  
import numpy as np # Create a DataFrame  
df = pd.DataFrame({
'A': [1, 4, 9],
'B': [16, 25, 36]})   #Apply square root function along columns  
result = df.apply(np.sqrt, axis=0)  
print(result)  
Explanation: Here, apply operates on entire rows or columns at a time, making it suitable for row/column-wise transformations or aggregations.  
applymap Method  
The applymap method is exclusive to DataFrames and applies a function element-wise to every individual value in the DataFrame.

 -  Key Differences  
 Scope: apply works on both Series and DataFrames, while applymap is limited to DataFrames.  
  Granularity: apply operates on rows, columns, or individual elements (depending on the axis), whereas applymap always works element-wise.

**Q21. Why is NumPy’s array slicing faster than Python’s list slicing?**

  -  NumPy's array slicing is faster than Python's list slicing due to several key reasons:  
  1.  Homogeneous Data and Contiguous Memory:  
  NumPy arrays store data in a contiguous block of memory and are homogeneous (all elements are of the same data type). This allows for efficient memory access and better cache utilization.
Python lists, on the other hand, are collections of pointers to objects, even if all elements are of the same type. This adds overhead during slicing as the interpreter has to dereference each pointer.  
2.  Vectorized Operations:  
NumPy is optimized for numerical computations and uses low-level C implementations for operations like slicing. This eliminates the need for Python's slower, interpreted loops.
Python lists rely on high-level Python loops, which are slower due to the overhead of dynamic type checking and object handling.  
3.  No Type Checking Overhead:  
Since NumPy arrays are homogeneous, there is no need for type checking during slicing. Python lists, being heterogeneous, require type checks for each element, which slows down the process.  
4.  Efficient Indexing:  
NumPy uses advanced indexing techniques that directly map to memory locations, making slicing operations extremely fast.
Python lists involve additional steps to handle pointers and objects, which increases the time complexity.
